```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt

# 定义函数：画出新数据点与 k 个最近邻的连线
def draw_neighbor(dogs, indexes, k, newdata):
    for i in range(k):
        neighbor_index = indexes[0][i]   # 第 i 个邻居的索引
        neighbor_point = dogs[neighbor_index]  # 邻居点坐标

        # 用虚线连接新数据点与邻居点
        plt.plot(
            [newdata[0][0], neighbor_point[0]],
            [newdata[0][1], neighbor_point[1]],
            'k--', linewidth=1
        )

        # 用黑色空心圆标出邻居点
        plt.scatter(
            neighbor_point[0], neighbor_point[1],
            s=150, facecolors='none', edgecolors='k',
            linewidths=1.5, label='Neighbor' if i==0 else ""
        )

# ======== 数据部分 ========

# 腊肠犬（Dachshund）的长度与高度数据
dach_length = [55, 57, 64, 63, 58, 49, 54, 61]
dach_height = [30, 31, 36, 30, 33, 25, 37, 34]

# 珍岛犬（Jindo dog）的长度与高度数据
jin_length = [56, 47, 56, 46, 49, 53, 52, 48]
jin_height = [52, 52, 50, 53, 50, 53, 49, 54]

# 构建数据与标签
j_data = np.column_stack((jin_length, jin_height))
j_label = np.ones(len(j_data))  # 1 表示珍岛犬

d_data = np.column_stack((dach_length, dach_height))
d_label = np.zeros(len(d_data))  # 0 表示腊肠犬

# 合并数据
dogs = np.concatenate((d_data, j_data))
labels = np.concatenate((d_label, j_label))

# 新数据（未知品种）
newdata = [[52, 42]]

dog_classes = {0: 'Dachshund（腊肠犬）', 1: 'Jindo dog（珍岛犬）'}

# ======== kNN 分类部分 ========

k = 5  # 可以改成 9 试试不同的结果
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(dogs, labels)

# 预测新数据的类别
y_pred = knn.predict(newdata)
print('新数据', newdata, '预测结果:', dog_classes[y_pred[0]])

# 找到最近的 k 个邻居
distances, indexes = knn.kneighbors(newdata)
print('距离:', distances)
print('邻居索引:', indexes)

# ======== 绘图部分 ========

plt.figure(figsize=(7,6))
plt.scatter(dach_length, dach_height, c='red', label='Dachshund（腊肠犬）')
plt.scatter(jin_length, jin_height, c='blue', marker='^', label='Jindo dog（珍岛犬）')
plt.scatter(newdata[0][0], newdata[0][1], s=100, marker='p', c='green', label='新数据')

# 画出与邻居的连接线
draw_neighbor(dogs, indexes, k, newdata)

plt.xlabel('Length (cm)')
plt.ylabel('Height (cm)')
plt.title(f'k = {k} 的 kNN 分类结果')
plt.leg
```
